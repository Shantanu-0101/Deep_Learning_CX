# Non efficient mehtod:

```python
#fibo # inefficient method/code
import time
def fib(n):
    if n==1 or n==0:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python

start=time.time()
fib(4)
print(time.time()-start)
```

```python
start=time.time()
fib(30)
print(time.time()-start)
```

```python
start=time.time()
fib(36)
print(time.time()-start)
```

# Efficient method/code

```python
d={0:1,1:1}
def fib(n,d):
    if n in d:
        return d[n]
    else:
        d[n] =  fib(n-1,d) + fib(n-2,d)
        return d[n]
```

```python
fib(30,d)
```

```python
fib(50,d)
```

```python
fib(100,d)
```
