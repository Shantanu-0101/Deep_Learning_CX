```python
from keras.layers import *
from keras.models import Model
```

```python
# Define two sets of Input

InputA = Input(shape=(32,))
InputB = Input(shape=(128,))
```

```python
# First branch operated on first input
x = Dense(8, activation='relu')(InputA)
x1 = Dense(4, activation='relu')(x)
```

```python
# Second branch operates on second input
y = Dense(64, activation='relu')(InputB)
y1 = Dense(32, activation='relu')(y)
y2 = Dense(16, activation='relu')(y1)
```

```python
# Combine the output of the two branches
combined = concatenate([x1,y1])
```

```python
# Apply a FC layer and then a regression prediction on the combined outputs
z = Dense(2, activation='relu')(combined)
z1 = Dense(1, activation='relu')(z)
```

```python
# our model will accept the inputs from two branches and output a single value
model = Model(inputs = [InputA,InputB], outputs = z)
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True)
```
